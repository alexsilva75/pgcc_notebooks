```python
BASE_PATH = '/home/alexsandro/dev/UEFS/PROJETO_PESQUISA/LACAD/pgcc-notebooks/data'

class_stain_count = {'1_Normal': {'AZAN': 97, 'HE': 716, 'PAMS': 156, 'PAS': 249},
                     '2_Esclerose_Pura_Sem_Crescente': {'AZAN': 234, 'HE': 672, 'PAMS': 104, 'PAS': 472},
                     '3_Hipercelularidade': {'AZAN': 93, 'HE': 653, 'PAS': 345},
                     '4_Hipercelularidade_Pura_Sem_Crescente': {'AZAN': 60,  'PAS': 164},
                     '5_Crescent': {'AZAN': 0, 'HE': 0, 'PAMS': 0, 'PAS': 0},
                     '6_Membranous': {'AZAN': 0, 'HE': 0, 'PAMS': 0, 'PAS': 0},
                     '7_Sclerosis': {'AZAN': 0, 'HE': 0, 'PAMS': 0, 'PAS': 0},
                     '8_Podocytopathy': {'AZAN': 0, 'HE': 0, 'PAMS': 0, 'PAS': 0},
                     }

def main():

    
    for class_folder in os.listdir(BASE_PATH):
        class_folder_path = str(BASE_PATH + os.sep + class_folder)

        for stain_folder in os.listdir(class_folder_path):
            stain_folder_path = str(class_folder_path + os.sep + stain_folder)

            for file_name in os.listdir(stain_folder_path):
                file_prefix = class_folder + '_'+ stain_folder
                extension = file_name.split('.')[1]
                if file_prefix not in file_name:
                    stain_file_count = class_stain_count[class_folder][stain_folder]
                    new_file_name = file_prefix + '_' + str(stain_file_count) + '.'+extension 
                    old_file_path = str(stain_folder_path + os.sep + file_name)
                    new_file_path = str(stain_folder_path + os.sep + new_file_name)
                    os.rename(old_file_path, new_file_path)
                    stain_file_count += 1
                    class_stain_count[class_folder][stain_folder] = stain_file_count
```

```python
main()
```
